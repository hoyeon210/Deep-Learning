### 활성화 함수(Activation fuction)

### 01. Sigmoid
- y = 1 / ( 1 + e<sup>-x</sup>)

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_function(x):
    return 1 / ( 1 + np.exp(-x))
```

```python
x = np.linspace(-5,5)
y = sigmoid_function(x)
plt.plot(x, y)
plt.title("Sigmoid")
plt.show()
```

```python
# sigmoid함수는 y값의 결과가 0과 1사이에서 나옴
```

### Sigmoid 함수 미분했을때
- y<sup>'</sup> = (1-y) * y

### 02. tanh
- y = (e<sup>x</sup> - e<sup>-x</sup>) / (e<sup>x</sup> + e<sup>-x</sup>)

```python
def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5,5)
y = tanh_function(x)
plt.plot(x, y)
plt.title("tanh")
plt.show()
```

```python
# tanh함수는 y값의 결과가 -1과 1사이에서 나옴
```

### 03. ReLU(Rectified Linear Unit)
- y = 0 if (x <= 0) else x

```python
def relu_function(x):
    return np.where(x <= 0, 0, x) 

x = np.linspace(-5,5)
y = relu_function(x)
plt.plot(x, y)
plt.title("ReLU")
plt.show()
```

### 04. Leaky ReLU
- X가 음수인 영역에서 아주 작은 기울기를 생성
- 출력이 0이 되어 더 이상 학습이 진행되지 않는 뉴런이 발생하는 dying ReLU 현상을 피하기 위해 사용
- y = 0.01 * x if(x <= 0) else x

```python
def leaky_relu_function(x):
    return np.where(x <= 0, 0.01*x, x) 

x = np.linspace(-5,5)
y = leaky_relu_function(x)
plt.plot(x, y)
plt.title("Leaky ReLU")
plt.show()
```

### 05.Softmax
- 입력받은 값을 출력으로 0 ~ 1 사이의 값으로 모두 정규화하며, 출력 값들의 총합은 항상 1이 되는 특성을 가진 함수
!

```python
def softmax_function(x):
    return np.exp(x) / np.sum(np.exp(x))

y = softmax_function(np.array([1,2,3]))
print(y)
```
